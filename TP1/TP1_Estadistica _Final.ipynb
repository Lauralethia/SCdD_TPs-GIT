# DEPURACIÓN DE DATOS

### A partir de los gráficos anteriores se detecta un valor atípico (outlier) para la condición de lluvia.Se realiza la depuración utilizando como umbral de eliminación 2 desvíos por sobre la media. A los valores detectados como outliers se les asgina NaN.

```python
#DETECCIÓN Y ELIMINACION DE OUTLIERS
#criterio de exclusión: se eliminarán aquellos valores por arriba o por debajo de dos desvíos estándar de la media de los datos
#por condición.

data = raw_data.copy()
for nr_cond in range(1,4):
    data_cond = data[:,nr_cond].copy()
    max_threshold = np.mean(data_cond) + 2 * np.std(data_cond)
    min_threshold = np.mean(data_cond) - 2 * np.std(data_cond)
    data_cond[data[:,nr_cond]>max_threshold] = np.nan
    data_cond[data[:,nr_cond]<min_threshold] = np.nan
    data[:,nr_cond] = data_cond
    
print(data)
```

### A continuación se grafican los datos curados:

```python
#VISUALIZACIÓN DATOS CURADOS, JUNTOS 
#(a partir de acá se trabajará con los datos curados solamente)
bins = np.linspace(0, 25, 10)

for i in range(1,4): 
    plt.hist(data[~np.isnan(data[:,i]),i],bins,alpha=0.5, label=cond_names[i-1])

plt.xlabel('Tiempo')
plt.ylabel('Cuentas') 
plt.legend(loc='upper right')
plt.show()

markers = ['o','x','+']

for i in range(1,4):     
    plt.scatter(data[~np.isnan(data[:,i]),0],data[~np.isnan(data[:,i]),i], label=cond_names[i-1],marker=markers[i-1])
        
plt.xlabel('Corredor')
plt.ylabel('Tiempo') 
plt.legend(loc='upper right')
plt.show()
```

# PRUEBAS DE NORMALIDAD E IGUALDAD DE VARIANZA

### Se realizan las pruebas de Shapiro-Wilks para evaluar normalidad de las variables y Levene para igualdad de varianzas.

```python
#PRUEBAS DE NORMALIDAD E IGUALDAD DE VARIANZA

#Para determinar qué tipo de análisis estadísticos se podrán utilizar, realizamos pruebas de normalidad (Shapiro Wilk) 
#e igualdad de varianza (F-test).

#Shapiro-Wilk test de normalidad
for i in range(1,4):         
    print(cond_names[i-1] + ": p = " + str(round(stats.shapiro(data[~np.isnan(data[:,i]),i])[0],4)) + " (Shapiro-Wilk).")    


#TODO CONSULTAR F-test para igualdad de varianzas 
#mientras Levene test
for i1 in range(1,4):
    for i2 in range(i1+1,4):
        res = stats.levene(data[~np.isnan(data[:,i1]),i1],data[~np.isnan(data[:,i2]),i2],center='mean')
        print(cond_names[i1-1] + ' vs ' + cond_names[i2-1] + ': p = ' + str(round(res[1],4)) + ' (Levene Test).')
        
     
```

### El test de Shapiro-Wilk no rechaza la hipótesis de normalidad de los datos para cada condición.
### El test Levene indica que la condición de lluvia tiene una varianza significativamente diferente a las de las otras dos condiciones.


```python
def exact_mc_perm_test2(xs, ys, nmc):
    n = len(xs)
    k=[]
    diff = np.mean(xs) - np.mean(ys)
    
    for j in range(nmc):
        zs = np.concatenate([xs, ys])
        np.random.shuffle(zs)
        k.append(np.mean(zs[:n]) - np.mean(zs[n:]))
    if diff > 0:
        return float(np.sum(k>diff))/float(nmc)
    else:
        return float(np.sum(k<diff))/float(nmc)
```

## Pregunta 1

### Hipótesis:  Los Atletas son mas lentos en dias de lluvia que en día soleado
###  Planteo: Lluvia < sol (test de una cola) - como las varianzas son diferentes se utiliza un test no paramétrico. Se implementa Rank Sum porque es mas adecuado para N<20

```python

h1=stats.ranksums(data[~np.isnan(data[:,1]),1],data[~np.isnan(data[:,3]),3] )#"It does not handle ties between measurements in x and y."

tval1=h1[0]
pval1=h1[1]
print('clima lluvia Vs Soleado p:' + str(pval1)+' (Rank Sum)')
    
pperm1=exact_mc_perm_test2(data[~np.isnan(data[:,1]),1],data[~np.isnan(data[:,3]),3], 100000) 
print('Comparacion clima lluvia Vs Soleado ' + str(pperm1)+ ' (permutaciones)')
```

### Conlusion: En clima de lluvia los atletas son mas lentos que en los dias de sol

## Pregunta 2

###  El cielo nublado no influye en los tiempos de los atletas

### Planteo: usamos la condición de día soleado como referencia. Dado que las variables ~N y tienen iguales varianzas, hacemos t Test relacionado

```python
#2) El cielo nublado no influye en los tiempos de los atletas
# Usamos dia Soleado como referencia. Dado que las variables ~N y tienen iguales varianzas, hacemos t Test relacionado

h2=stats.ttest_rel(data[~np.isnan(data[:,1]),1],data[~np.isnan(data[:,2]),2])
tval2=h2[0] 
pval2=h2[1]
print('clima nublado Vs Soleado p:' + str(round(pval2,4))+' (ttest_rel)')

    
pperm2=exact_mc_perm_test2(data[:,1],data[:,2], 1000)
print('Comparacion clima nublado Vs Soleado ' + str(pperm2)+ ' (permutaciones)')
# Conlusion: El dia nublado no influye en el tiempo de los atletas respecto de los dias de sol
```

### Conlusion: El dia nublado no influye en el tiempo de los atletas respecto de los dias de sol
