```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from  scipy import stats
%matplotlib inline
plt.rcParams.update({'figure.figsize': (10.0, 6.0), 'font.size': 18})

#load file
file_name = 'tiempos.txt'

raw_data = np.loadtxt(file_name,skiprows=1)
```

## INSPECCION DE DATOS CRUDOS
* SEPARADO POR CONDICION


```python
#INSPECCION DE DATOS CRUDOS
cond_names=["Sol","Nublado","Lluvia"]   

#SEPARADO POR CONDICION
f, axarr = plt.subplots(3, 2) 
for i in range(1,4):
    axarr[i-1,0].hist(raw_data[:,i])    
    axarr[i-1,1].scatter(raw_data[:,0],raw_data[:,i])
    axarr[i-1,0].set_ylabel(cond_names[i-1])
    
axarr[2,0].set_xlabel('Tiempo')
axarr[2,1].set_xlabel('Corredor')    

plt.show()
```

```python
#VISUALIZACIÓN DATOS CRUDOS JUNTOS
bins = np.linspace(0, 220, 100)

for i in range(1,4): 
    plt.hist(raw_data[:,i],bins,alpha=0.5, label=cond_names[i-1])

plt.xlabel('Tiempo')
plt.ylabel('Cuentas') 
plt.legend(loc='upper right')
plt.show()

markers = ['o','x','+']

for i in range(1,4):     
    plt.scatter(raw_data[:,0],raw_data[:,i], label=cond_names[i-1],marker=markers[i-1])


plt.xlabel('Corredor')
plt.ylabel('Tiempo') 
plt.legend(loc='upper right')
plt.show()

print('Por inspección visual se observa que hay un outlier en el caso de lluvia.')
```

```python
#DETECCIÓN Y ELIMINACION DE OUTLIERS
#criterio de exclusión: se eliminarán aquellos valores por arriba o por debajo de dos desvíos estándar de la media de los datos
#por condición.

data = raw_data.copy()
for nr_cond in range(1,4):
    data_cond = data[:,nr_cond].copy()
    max_threshold = np.mean(data_cond) + 2 * np.std(data_cond)
    min_threshold = np.mean(data_cond) - 2 * np.std(data_cond)
    data_cond[data[:,nr_cond]>max_threshold] = np.nan
    data_cond[data[:,nr_cond]<min_threshold] = np.nan
    data[:,nr_cond] = data_cond
    
print(data)
print('Siguiendo este criterio se asigna como nan a un único outlier, que fue observado durante la insepcción.')
```

```python
#VISUALIZACIÓN DATOS CURADOS, JUNTOS 
#(a partir de acá se trabajará con los datos curados solamente)
bins = np.linspace(0, 25, 10)

for i in range(1,4): 
    plt.hist(data[~np.isnan(data[:,i]),i],bins,alpha=0.5, label=cond_names[i-1])

plt.xlabel('Tiempo')
plt.ylabel('Cuentas') 
plt.legend(loc='upper right')
plt.show()

markers = ['o','x','+']

for i in range(1,4):     
    plt.scatter(data[~np.isnan(data[:,i]),0],data[~np.isnan(data[:,i]),i], label=cond_names[i-1],marker=markers[i-1])
        
plt.xlabel('Corredor')
plt.ylabel('Tiempo') 
plt.legend(loc='upper right')
plt.show()
```

```python
#PRUEBAS DE NORMALIDAD E IGUALDAD DE VARIANZA

#Para determinar qué tipo de análisis estadísticos se podrán utilizar, realizamos pruebas de normalidad (Shapiro Wilk) 
#e igualdad de varianza (F-test).

#Shapiro-Wilk test de normalidad
for i in range(1,4):         
    print(cond_names[i-1] + ": p = " + str(round(stats.shapiro(data[~np.isnan(data[:,i]),i])[0],4)) + " (Shapiro-Wilk).")    


#TODO CONSULTAR F-test para igualdad de varianzas 
#mientras Levene test
for i1 in range(1,4):
    for i2 in range(i1+1,4):
        res = stats.levene(data[~np.isnan(data[:,i1]),i1],data[~np.isnan(data[:,i2]),i2],center='mean')
        print(cond_names[i1-1] + ' vs ' + cond_names[i2-1] + ': p = ' + str(round(res[1],4)) + ' (Levene Test).')
        
#El test de Shapiro-Wilk no rechaza la hipótesis de normalidad de los datos para cada condición.
#El test Levene indica que la condición de lluvia tiene una varianza significativamente diferente a las de las otras dos 
#condiciones. 
#Conclusión: Se podrá utilizar estadística paramétrica teniendo en cuenta la heterogeneidad de varianzas.        
```

```python
def exact_mc_perm_test2(xs, ys, nmc):
    n = len(xs)
    k=[]
    diff = np.mean(xs) - np.mean(ys)
    
    for j in range(nmc):
        zs = np.concatenate([xs, ys])
        np.random.shuffle(zs)
        k.append(np.mean(zs[:n]) - np.mean(zs[n:]))
    if diff > 0:
        return float(np.sum(k>diff))/float(nmc)
    else:
        return float(np.sum(k<diff))/float(nmc)
```

```python
#Abordaje de las preguntas
#1) Los Atletas son mas lentos en dias de lluvia que en día soleado
# otra posible sol, restar y welsh contra 0
# lluvia < sol (test de una cola) - no paramétrico. Se implementa Rank Sum por N<20
h1=stats.ranksums(data[~np.isnan(data[:,1]),1],data[~np.isnan(data[:,3]),3] )#"It does not handle ties between measurements in x and y."

tval1=h1[0]
pval1=h1[1]
print('clima lluvia Vs Soleado p:' + str(pval1)+' (Rank Sum)')
    
pperm1=exact_mc_perm_test2(data[~np.isnan(data[:,1]),1],data[~np.isnan(data[:,3]),3], 100000) 
print('Comparacion clima lluvia Vs Soleado ' + str(pperm1)+ ' (permutaciones)')
# Conlusion: En clima de lluvia los atletas son mas lentos que en los dias de sol
```

```python
#2) El cielo nublado no influye en los tiempos de los atletas
# Usamos dia Soleado como referencia. Dado que las variables ~N y tienen iguales varianzas, hacemos t Test relacionado

h2=stats.ttest_rel(data[~np.isnan(data[:,1]),1],data[~np.isnan(data[:,2]),2])
tval2=h2[0] 
pval2=h2[1]
print('clima nublado Vs Soleado p:' + str(round(pval2,4))+' (ttest_rel)')

    
pperm2=exact_mc_perm_test2(data[:,1],data[:,2], 1000)
print('Comparacion clima nublado Vs Soleado ' + str(pperm2)+ ' (permutaciones)')
# Conlusion: El dia nublado no influye en el tiempo de los atletas respecto de los dias de sol
```

```python
#3) La velocidad en días de lluvia es independiente del atleta

#puedo comparar varianzas? Supongo que en la lluvia todos hacen tiempos parecidos
h3_a=stats.pearsonr(data[:,1],data[:,2])
rval3_a=h3_a[0] 
pval3_a=h3_a[1]

h3_b=stats.pearsonr(data[:,1],data[:,3])
rval3_b=h3_b[0] 
pval3_b=h3_b[1]
    
h3_c=stats.pearsonr(data[:,2],data[:,3])
rval3_c=h3_c[0] 
pval3_c=h3_c[1]
print('correlaciones de pearson, atletas en buen clima ..')
print('ScN: ' + str(pval3_a))
print('...y atletas en lluvia respecto a desempenio en cualquier clima bueno')
print('LLcS: ' + str(pval3_b))
print('LLcN: ' + str(pval3_c))

## comparar grupos definidos en sol, en lluvia

## El modelo de que los atletas performan igual independientemente del clima se rompe los dias de lluvia. Si bien la correlacion es 
## en dias de sol y nublados se encuentra que la correlacion con dias de lluvia es baja. Sin embargo esto no alcanza para concluir que 
## en dias de lluvia la velocidad es independiente delos atletas
```

```python
#4) El clima influye en la velocidad de los atletas.

h4=stats.f_oneway(data[:,1],data[:,2],data[~np.isnan(data[:,3]),3])
tval4=h4[0]
pval4=h4[1]
print('Anova-uneway entre climas p:' + str(round(h4,4)))
```
